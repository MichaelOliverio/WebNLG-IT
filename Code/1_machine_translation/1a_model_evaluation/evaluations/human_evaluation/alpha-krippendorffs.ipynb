## Alpha di Krippendorffs

```python
# read p annotations
df_p = pd.read_excel(f'{folder_path}/{file_name}', sheet_name='1_p')

# mantieni solamente la colonna 2 e 3
df_p = df_p[['Adequacy (1-5)', 'Fluency (1-5)']]

# rimuovi la prima riga
df_p = df_p.iloc[1:]

# ripristina l'indice
df_p = df_p.reset_index(drop=True)

# ottieni solo le prime 100 righe
df_p = df_p.iloc[:100]

df_p
```

```python
# read michael annotations
df_michael = pd.read_excel(f'{folder_path}/{file_name}', sheet_name='2_Michael')

# mantieni solamente la colonna 2 e 3
df_michael = df_michael[['Adequacy (1-5)', 'Fluency (1-5)']]

# rimuovi la prima riga
df_michael = df_michael.iloc[1:]

# ripristina l'indice
df_michael = df_michael.reset_index(drop=True)

# ottieni solo le prime 100 righe
df_michael = df_michael.iloc[:100]

df_michael
```

```python
# Unisco le annotazioni
unit_ids = list(range(len(df_p)))
annotator1_id = 'P'
annotator2_id = 'M'

# Creo un DataFrame combinato
combined_data = []
for i in unit_ids:
    combined_data.append([i, annotator1_id, df_p.loc[i, 'Fluency (1-5)'], df_p.loc[i, 'Adequacy (1-5)']])
    combined_data.append([i, annotator2_id, df_michael.loc[i, 'Fluency (1-5)'], df_michael.loc[i, 'Adequacy (1-5)']])

combined_df = pd.DataFrame(combined_data, columns=['unit_id', 'annotator_id', 'Fluency (1-5)', 'Adequacy (1-5)'])
combined_df.head()
```

```python
import simpledorff

# Calcolo dell'alpha di Krippendorff per 'fluency'
alpha_fluency = simpledorff.calculate_krippendorffs_alpha_for_df(
    combined_df,
    experiment_col='unit_id',
    annotator_col='annotator_id',
    class_col='Fluency (1-5)'
)

# Calcolo l'alpha di Krippendorff per 'adequacy'
alpha_adequacy = simpledorff.calculate_krippendorffs_alpha_for_df(
    combined_df,
    experiment_col='unit_id',
    annotator_col='annotator_id',
    class_col='Adequacy (1-5)'
)

print(f"Krippendorff's alpha per fluency: {alpha_fluency}")
print(f"Krippendorff's alpha per adequacy: {alpha_adequacy}")
```
