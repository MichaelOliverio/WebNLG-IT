```python
import pandas as pd
import numpy as np
import os
import xml.etree.ElementTree as ET
from tqdm import tqdm
```

```python
# read the csv file
df = pd.read_csv('../generations/it5-large_it_2epoch_decoding.csv')
df['data points'] = np.nan
df
```

```python
path = "../../../../WebNLG/it-PE/test"

# Ciclo basato sugli indici del DataFrame
for i in range(len(df)):
    actual = df.at[i, 'actuals']
    data_points = []
    flag = False
    print(f"Entry {i + 1}")
    print(f"Actual: {actual}")
    
    for file_name in os.listdir(path):
        if os.path.isfile(os.path.join(path, file_name)):
            url = os.path.join(path, file_name)
            tree = ET.parse(url)
            root = tree.getroot()
            
            for entry in root.iter('entry'):
                for lex in entry.iter('lex'):
                    if lex.attrib['lang'] == 'it-PE':
                        data_points.append(lex.text)
                        if lex.text == actual:
                            flag = True
                if flag:
                    break
                else:
                    data_points = []

            if flag:
                print(f"{data_points}")
                print('\n')

                # Aggiungi la lista come singolo elemento nel DataFrame per la riga specifica
                df.at[i, 'data points'] = data_points
                break
df
```

```python
#save the cols "data_points" into csv file
df['data points'].to_csv('data_points.csv', index=False)
```
