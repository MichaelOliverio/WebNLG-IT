## Alpha di Krippendorffs LLaMA

```python
# read p annotations
df_p = pd.read_excel(f'{folder_path}/{file_name}', sheet_name='p-LLaMA')

# mantieni solamente la colonna 2 e 3
df_p = df_p[['#INCORRECT', '#NOT_CHECKABLE', '#MISLEADING', '#OTHER']]

# ripristina l'indice
df_p = df_p.reset_index(drop=True)

# ottieni solo le prime 50 righe
df_p = df_p.iloc[:50]

# read p annotations
df_michael = pd.read_excel(f'{folder_path}/{file_name}', sheet_name='m-LLaMA')

# mantieni solamente la colonna 2 e 3
df_michael = df_michael[['#INCORRECT', '#NOT_CHECKABLE', '#MISLEADING', '#OTHER']]

# ripristina l'indice
df_michael = df_michael.reset_index(drop=True)

# ottieni solo le prime 50 righe
df_michael = df_michael.iloc[:50]

# Unisco le annotazioni
unit_ids = list(range(len(df_p)))
annotator1_id = 'P'
annotator2_id = 'M'

# Creo un DataFrame combinato
combined_data = []
for i in unit_ids:
    combined_data.append([i, annotator1_id, df_p.loc[i, '#INCORRECT'], df_p.loc[i, '#NOT_CHECKABLE'], df_p.loc[i, '#MISLEADING'], df_p.loc[i, '#OTHER']])
    combined_data.append([i, annotator2_id, df_michael.loc[i, '#INCORRECT'], df_michael.loc[i, '#NOT_CHECKABLE'], df_michael.loc[i, '#MISLEADING'], df_michael.loc[i, '#OTHER']])

combined_df = pd.DataFrame(combined_data, columns=['unit_id', 'annotator_id', '#INCORRECT', '#NOT_CHECKABLE', '#MISLEADING', '#OTHER'])


# Calcolo dell'alpha di Krippendorff per 'INCORRECT'
alpha_incorrect = simpledorff.calculate_krippendorffs_alpha_for_df(
    combined_df,
    experiment_col='unit_id',
    annotator_col='annotator_id',
    class_col='#INCORRECT'
)

# Calcolo l'alpha di Krippendorff per 'NOT_CHECKABLE'
alpha_not_checkable = simpledorff.calculate_krippendorffs_alpha_for_df(
    combined_df,
    experiment_col='unit_id',
    annotator_col='annotator_id',
    class_col='#NOT_CHECKABLE'
)

# Calcolo l'alpha di Krippendorff per 'MISLEADING'
alpha_misleading = simpledorff.calculate_krippendorffs_alpha_for_df(
    combined_df,
    experiment_col='unit_id',
    annotator_col='annotator_id',
    class_col='#MISLEADING'
)

# Calcolo l'alpha di Krippendorff per 'OTHER'
alpha_other = simpledorff.calculate_krippendorffs_alpha_for_df(
    combined_df,
    experiment_col='unit_id',
    annotator_col='annotator_id',
    class_col='#OTHER'
)
print(f"Krippendorff's alpha per LLaMA incorrect: {alpha_incorrect}")
print(f"Krippendorff's alpha per LLaMA not checkable: {alpha_not_checkable}")
print(f"Krippendorff's alpha per LLaMA misleading: {alpha_misleading}")
print(f"Krippendorff's alpha per LLaMA other: {alpha_other}")

total = alpha_incorrect + alpha_not_checkable + alpha_misleading + alpha_other
```
